##Functions and imports (Run once at start and when you add/change functions)

````python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from PIL import Image
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator,img_to_array, load_img
import os
from tensorflow.keras.callbacks import EarlyStopping,ModelCheckpoint
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, RandomRotation, RandomZoom, RandomFlip, GaussianNoise
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.losses import CategoricalCrossentropy


#Defined functions
import numpy as np
def plot_history(history):
  plt.figure(figsize=(14, 5))
  plt.subplot(1, 2, 1)
  plt.plot(history.history['accuracy'], label='Train')
  plt.plot(history.history['val_accuracy'], label='Validation')
  plt.title('Model accuracy')
  plt.ylabel('Accuracy')
  plt.xlabel('Epoch')
  plt.legend(loc='upper left')

  # Plot training loss
  plt.subplot(1, 2, 2)
  plt.plot(history.history['loss'], label='Train')
  plt.plot(history.history['val_loss'], label='Validation')
  plt.title('Model loss')
  plt.ylabel('Loss')
  plt.xlabel('Epoch')
  plt.legend(loc='upper left')

  plt.show()


def preprocess_image(image_path, target_size=(48, 48)):
  """Load an image and preprocess it for emotion classification."""
  img = load_img(image_path, target_size=target_size, color_mode='rgb')
  img_array = img_to_array(img)
  img_array_expanded_dims = np.expand_dims(img_array, axis=0)
  return img_array_expanded_dims

emotion_labels = ['angry', 'fear', 'happy', 'sad', 'surprise', 'neutral']

def classify_emotion(image_path):
  preprocessed_image = preprocess_image(image_path)
  prediction = model.predict(preprocessed_image)
  predicted_class = np.argmax(prediction)
  return emotion_labels[predicted_class]

def display_prediction(image_path):
  emotion = classify_emotion(image_path)
  img = plt.imread(image_path)
  plt.imshow(img)
  plt.title(f'Predicted Emotion: {emotion}')
  plt.axis('off')
  plt.show()

````

## VARS

````python
#VARS
BATCH_SIZE = 128 # BATCH SIZE
EPOCHS = 50 # EPOCHS
EARLY_STOP = 10 # SET THE EARLY STOP NUMBER FOR CALLBACK
SUBSET_FRACT = 1 # SCALES THE BATCH SIZE, (ex: 128 * 0.5 = 256, batch size = 256)

````

## DATA LOADING and DATA GENS

````python
# Initialize
datagen = ImageDataGenerator(validation_split=0.2)
data_dir = '/content/augmented/all_data'

# Prepare Generators
train_generator = datagen.flow_from_directory(
    data_dir,
    target_size=(48, 48),
    batch_size=int(BATCH_SIZE * SUBSET_FRACT),  # Adjusted batch size
    class_mode='categorical',
    subset='training',
    shuffle=True,
    seed=42
)

validation_generator = datagen.flow_from_directory(
    data_dir,
    target_size=(48, 48),
    batch_size=int(BATCH_SIZE * SUBSET_FRACT),  # Adjusted batch size
    class_mode='categorical',
    subset='validation',
    shuffle=True,
    seed=42
)
````

## MODEL STUFF


````python
models = {}
````

### Model #1 - Simple_cnn

````python

models['simple_cnn'] = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(48, 48, 3)),
    MaxPooling2D(2),
    Conv2D(64, (3), activation='relu'),
    MaxPooling2D(2, 2),
    Conv2D(128, (3), activation='relu'),
    MaxPooling2D(2),
    Flatten(),
    Dense(512, activation='relu'),
    Dropout(0.3),
    Dense(len(train_generator.class_indices), activation='softmax')
])


models['simple_cnn'].compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
````

### Model #2 - simple_cnn_LR

````python

models['simple_cnn_LR'] = Sequential([
    Conv2D(128, (3,3), activation='relu'),
    MaxPooling2D(2),
    Conv2D(256, (3,3), activation='relu'),
    MaxPooling2D(2),
    Conv2D(512, (3,3), activation='relu'),
    MaxPooling2D(2),
    Flatten(),

    Dense(128, activation='relu'),
    Dropout(0.2),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(len(train_generator.class_indices), activation='softmax')
])

models['simple_cnn_LR'].compile(
    optimizer=Adam(learning_rate=0.0005),
    loss=CategoricalCrossentropy(),
    metrics=['accuracy']
)
````

model 3: modified_cnn

````python
models['modified_cnn'] = Sequential([
    Conv2D(128, (3,3), activation='relu'),
    MaxPooling2D(2),
    Conv2D(256, (3,3), activation='relu'),
    MaxPooling2D(2),
    Conv2D(512, (3,3), activation='relu'),
    MaxPooling2D(2),
    Conv2D(512, (3,3), activation='relu'),  # Additional Conv2D layer
    MaxPooling2D(2),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.2),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(len(train_generator.class_indices), activation='softmax')
])

# Compile the model
models['modified_cnn'].compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
````

````python

models['high_dropout_LR'] = Sequential([
    Conv2D(128, (3,3), activation='relu'),
    MaxPooling2D(2),
    Conv2D(256, (3,3), activation='relu'),
    MaxPooling2D(2),
    Conv2D(512, (3,3), activation='relu'),
    MaxPooling2D(2),
    Flatten(),

    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(len(train_generator.class_indices), activation='softmax')
])

models['high_dropout_LR'].compile(
    optimizer=Adam(learning_rate=0.0001),
    loss=CategoricalCrossentropy(),
    metrics=['accuracy']
)
````

### Callbacks


*   Early Stopping
*   Check pointing



````python
from datetime import datetime
import os


# Define the checkpoint directory
checkpoint_dir = '/content/checkpoints'
os.makedirs(checkpoint_dir, exist_ok=True)

# Set up the checkpoint callback to save only the best model based on validation accuracy


model_checkpoint = ModelCheckpoint(
    filepath=os.path.join(checkpoint_dir, '4.h5'),
    save_best_only=True,
    monitor='val_accuracy',
    mode='max',
    verbose=1
)


callback_list = [
    EarlyStopping(monitor="val_accuracy", patience=EARLY_STOP, restore_best_weights=True),
    EarlyStopping(monitor="val_loss", patience=EARLY_STOP, restore_best_weights=True),
    ReduceLROnPlateau(factor=0.8, monitor="val_accuracy", patience=3),
    model_checkpoint
]
````

## FIT AND TRAIN


````python

# Now fit the model with the checkpoint callback included
history = models['simple_cnn'].fit(
    train_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    epochs=EPOCHS,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // validation_generator.batch_size,
    callbacks=callback_list,
    verbose=2
)
````
