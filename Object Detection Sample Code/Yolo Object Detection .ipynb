```python
from ultralytics import YOLO
import cv2
import math 
# start webcam
capture = cv2.VideoCapture(0)
capture.set(3, 640)
capture.set(4, 480)

# model
model = YOLO("yolo-Weights/yolov8n.pt")

# object classes
classNames = ["person", "bicycle", "car", "motorbike", "aeroplane", "bus", "train", "truck", "boat",
              "traffic light", "fire hydrant", "stop sign", "parking meter", "bench", "bird", "cat",
              "dog", "horse", "sheep", "cow", "elephant", "bear", "zebra", "giraffe", "backpack", "umbrella",
              "handbag", "tie", "suitcase", "frisbee", "skis", "snowboard", "sports ball", "kite", "baseball bat",
              "baseball glove", "skateboard", "surfboard", "tennis racket", "bottle", "wine glass", "cup",
              "fork", "knife", "spoon", "bowl", "banana", "apple", "sandwich", "orange", "broccoli",
              "carrot", "hot dog", "pizza", "donut", "cake", "chair", "sofa", "pottedplant", "bed",
              "diningtable", "toilet", "tvmonitor", "laptop", "mouse", "remote", "keyboard", "cell phone",
              "microwave", "oven", "toaster", "sink", "refrigerator", "book", "clock", "vase", "scissors",
              "teddy bear", "hair drier", "toothbrush"
              ]


while True:
    retrieved, image = capture.read()
    #here retrieved is a boolean output which tells us if the frame is read correctly or not and 
    #img is the frame itself. capture.read returns a tuple with these two values
    
    results = model(image, stream=True)
    #here the frome/image is passed through the model for object detection and stream = True means the the input 
    #is a video instead of a still image
    
    # coordinates for the detected objects
    for i in results:    #results contains the detected objects from the current frame which is itereated over
        boxes = i.boxes  #with i to extract the objects. Each object contains the info about the bounding boxes

        for box in boxes:
            #Iterating over all the objects one by one and getting the coordinates for it through box.xyxy[0]
            x1, y1, x2, y2 = box.xyxy[0]
            x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2) # converting the coordinates into integer values

            # putting the image into a bounding box with color and thickness specifications 
            cv2.rectangle(image, (x1, y1), (x2, y2), (255, 3, 255), 2)

            # Displaying the confidence level of the detected object
            confidence = math.ceil((box.conf[0]*100))/100
            print("Confidence --->",confidence)

            # Displaying the class of the detected object
            cls = int(box.cls[0])
            print("Class name -->", classNames[cls])

            # Specifing the coordinates, font and color to display the information over the bounding box
            org = [x1, y1]
            font = cv2.FONT_HERSHEY_DUPLEX
            fontScale = 1
            color = (255, 0, 0)
            thickness = 2
            
            #using the above specifications
            cv2.putText(image, classNames[cls], org, font, fontScale, color, thickness)

    cv2.imshow('Webcam', image)
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
